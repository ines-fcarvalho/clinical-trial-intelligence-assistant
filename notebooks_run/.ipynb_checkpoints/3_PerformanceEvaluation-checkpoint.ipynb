```python
# ================================
# Evaluate all HINT models on CTOD reduced test data
# ================================
from pathlib import Path
import joblib, numpy as np
from scipy import sparse as sp
import xgboost as xgb

# -------- CONFIG --------
# -------- CONFIG --------

EXPORT_CSV = False  # set True to export per-phase CSVs
EXPORT_DIR = Path("./quick_metric_tables")  # used if EXPORT_CSV=True
EXPORT_DIR.mkdir(parents=True, exist_ok=True)

PHASES = ["phase_I", "phase_II", "phase_III"]
MODEL_BASE = Path("/Users/antoniocortes/Tese/MyModel(hybrid)/xgb_model_package")
CTOD_ARTIFACTS = Path("/Users/antoniocortes/Tese/MyModel(hybrid)/ctod_xgb_artifacts")
HINT_ARTIFACTS = Path("/Users/antoniocortes/Tese/MyModel(hybrid)/hint_xgb_artifacts")

from sklearn.metrics import (
    roc_auc_score, average_precision_score, accuracy_score,
    precision_recall_fscore_support, confusion_matrix, roc_curve, precision_recall_curve
)
import numpy as np

def metrics_at_threshold(y_true, proba, thr=0.5):
    pred = (proba >= thr).astype(int)
    acc  = accuracy_score(y_true, pred)
    prec, rec, f1, _ = precision_recall_fscore_support(y_true, pred, average="binary", zero_division=0)
    tn, fp, fn, tp = confusion_matrix(y_true, pred).ravel()
    tnr = tn / (tn + fp + 1e-9)     # specificity
    bal = 0.5 * (tnr + rec)         # balanced accuracy
    return dict(threshold=float(thr), accuracy=float(acc), precision=float(prec), recall=float(rec),
                f1=float(f1), tnr=float(tnr), bal_acc=float(bal), cm=dict(tn=int(tn), fp=int(fp), fn=int(fn), tp=int(tp)))

def pick_threshold_balacc_max(y_true, proba):
    fpr, tpr, thr = roc_curve(y_true, proba)
    # Add 0.5 as a reference candidate
    candidates = np.unique(np.r_[thr, 0.5])
    def balacc_at(t):
        pred=(proba>=t).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_true, pred).ravel()
        tnr = tn/(tn+fp+1e-9); tpr_=tp/(tp+fn+1e-9)
        return 0.5*(tnr+tpr_)
    scores = np.array([balacc_at(t) for t in candidates])
    idx = int(np.argmax(scores))
    return float(candidates[idx]), float(scores[idx])

def pick_threshold_tnr_target(y_true, proba, target_tnr=0.80):
    fpr, tpr, thr = roc_curve(y_true, proba)
    tnr = 1.0 - fpr
    ok = np.where(tnr >= target_tnr)[0]
    if len(ok):
        i = ok[-1]  # lowest thr that still satisfies target
    else:
        i = int(np.lexsort((-tpr, tnr))[-1])  # best feasible
    return float(thr[i]), dict(tnr=float(tnr[i]), tpr=float(tpr[i]))

# -------- HELPERS --------
def load_ctod_test(phase: str):
    """Load CTOD joined (train+valid) test matrix reduced by its SVD."""
    pdir = CTOD_ARTIFACTS / phase
    X = sp.load_npz(pdir / "X_test_joined_reduced.npz")
    y = np.load(pdir / "y_test_joined.npy")
    return X, y

def load_hint_model(phase: str):
    """Load HINT-trained XGBoost model (joblib or fallback JSON)."""
    mdir = MODEL_BASE / phase
    joblib_path = mdir / "hint_xgb_model.joblib"
    json_path   = mdir / "xgb_model.json"
    if joblib_path.exists():
        return joblib.load(joblib_path)
    booster = xgb.Booster()
    booster.load_model(str(json_path))
    mdl = xgb.XGBClassifier()
    mdl._Booster = booster
    return mdl

# -------- MAIN LOOP --------
results = {}

for phase in PHASES:
    print(f"\n=== {phase.upper()} ===")
    X_test, y_test = load_ctod_test(phase)
    model = load_hint_model(phase)
    proba_test = model.predict_proba(X_test)[:, 1]

    pos, neg = (y_test == 1).sum(), (y_test == 0).sum()
    print(f"Loaded HINT model on CTOD {phase} | "
          f"X_test={X_test.shape}, positives={pos}, negatives={neg}")

    results[phase] = {
        "y_true": y_test,
        "y_proba": proba_test,
        "positives": int(pos),
        "negatives": int(neg)
    }

print("\nAll phases processed successfully!")



# ================================
# Evaluate all HINT models on HINT reduced test data
# ================================
from pathlib import Path
import joblib, numpy as np
from scipy import sparse as sp
import xgboost as xgb



# -------- METRICS (optional helpers kept for later use) --------
from sklearn.metrics import (
    roc_auc_score, average_precision_score, accuracy_score,
    precision_recall_fscore_support, confusion_matrix, roc_curve, precision_recall_curve
)

def metrics_at_threshold(y_true, proba, thr=0.5):
    pred  = (proba >= thr).astype(int)
    acc   = accuracy_score(y_true, pred)
    prec, rec, f1, _ = precision_recall_fscore_support(y_true, pred, average="binary", zero_division=0)
    tn, fp, fn, tp = confusion_matrix(y_true, pred).ravel()
    tnr   = tn / (tn + fp + 1e-9)
    bal   = 0.5 * (tnr + rec)
    return dict(threshold=float(thr), accuracy=float(acc), precision=float(prec), recall=float(rec),
                f1=float(f1), tnr=float(tnr), bal_acc=float(bal),
                cm=dict(tn=int(tn), fp=int(fp), fn=int(fn), tp=int(tp)))

def pick_threshold_balacc_max(y_true, proba):
    fpr, tpr, thr = roc_curve(y_true, proba)
    candidates = np.unique(np.r_[thr, 0.5])
    def balacc_at(t):
        pred=(proba>=t).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_true, pred).ravel()
        tnr = tn/(tn+fp+1e-9); tpr_=tp/(tp+fn+1e-9)
        return 0.5*(tnr+tpr_)
    scores = np.array([balacc_at(t) for t in candidates])
    i = int(np.argmax(scores))
    return float(candidates[i]), float(scores[i])

# -------- HELPERS --------
def _first_existing(*paths: Path) -> Path:
    for p in paths:
        if p.exists():
            return p
    raise FileNotFoundError("None of the expected artifact paths were found:\n" + "\n".join(map(str, paths)))

def load_hint_test(phase: str):
    """
    Load HINT test split reduced by SVD.
    Tries common layouts automatically.
    """
    # Try both directory layouts
    base1 = HINT_ARTIFACTS / phase
    base2 = HINT_ARTIFACTS / "phases_reduced" / phase

    X_path = _first_existing(base1 / "X_test_reduced.npz",
                             base2 / "X_test_reduced.npz")
    y_path = _first_existing(base1 / "y_test.npy",
                             base2 / "y_test.npy")

    X = sp.load_npz(X_path)
    y = np.load(y_path)
    return X, y

def load_hint_model(phase: str):
    """Load HINT-trained XGBoost model (joblib or fallback JSON)."""
    mdir = MODEL_BASE / phase
    joblib_path = mdir / "hint_xgb_model.joblib"
    json_path   = mdir / "xgb_model.json"
    if joblib_path.exists():
        return joblib.load(joblib_path)
    booster = xgb.Booster()
    booster.load_model(str(json_path))
    mdl = xgb.XGBClassifier()
    mdl._Booster = booster
    return mdl

# -------- MAIN LOOP --------
results = {}

for phase in PHASES:
    print(f"\n=== {phase.upper()} ===")
    X_test, y_test = load_hint_test(phase)
    model = load_hint_model(phase)
    proba_test = model.predict_proba(X_test)[:, 1]

    pos, neg = int((y_test == 1).sum()), int((y_test == 0).sum())
    print(f"Loaded HINT model on HINT {phase} | X_test={X_test.shape}, positives={pos}, negatives={neg}")

    results[phase] = {
        "y_true": y_test,
        "y_proba": proba_test,
        "positives": pos,
        "negatives": neg
    }

print("\nAll phases processed successfully on HINT test data!")
```

```python
# ===============================================
# Summary metrics (PR | F1 | ROC) for CTOD and HINT
# Displays: CTOD table, HINT table, and a combined side-by-side table
# ===============================================
import numpy as np
import pandas as pd
from IPython.display import display, Markdown
from sklearn.metrics import precision_recall_curve, average_precision_score, roc_auc_score

def best_f1_from_pr(y_true, proba):
    prec, rec, ths = precision_recall_curve(y_true, proba)
    f1s = 2 * prec * rec / (prec + rec + 1e-9)
    # ignore the last PR point (no threshold); guard against all-NaN
    core = f1s[:-1]
    return float(np.nanmax(core)) if np.isfinite(np.nanmax(core)) else np.nan

def summary_for_dataset(phases, load_xy_func, load_model_func):
    rows = []
    for phase in phases:
        X, y = load_xy_func(phase)
        model = load_model_func(phase)
        proba = model.predict_proba(X)[:, 1]

        ap  = average_precision_score(y, proba)
        roc = roc_auc_score(y, proba)
        f1b = best_f1_from_pr(y, proba)

        rows.append({
            "Phase": phase,
            "PR (Average Precision)": ap,
            "F1 (Best)": f1b,
            "ROC AUC": roc,
        })
    return pd.DataFrame(rows).set_index("Phase")[["PR (Average Precision)", "F1 (Best)", "ROC AUC"]]

# --- Build tables ---
df_ctod = summary_for_dataset(PHASES, load_ctod_test, load_hint_model)
df_hint = summary_for_dataset(PHASES, load_hint_test, load_hint_model)



# --- Combined side-by-side table ---
# Align indices and concatenate with a column MultiIndex
df_ctod_aligned = df_ctod.reindex(PHASES)
df_hint_aligned = df_hint.reindex(PHASES)
combined = pd.concat(
    {"CTOD": df_ctod_aligned, "HINT": df_hint_aligned},
    axis=1
)

display(Markdown("### Summary — side by side (CTOD vs HINT)"))
display(combined.style.format("{:.4f}"))




```

```python
# ===============================================
# Confusion Matrices — Rows: Phases | Columns: Datasets (CTOD, HINT)
# ===============================================
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score

def plot_confusion_matrices_ctod_and_hint_grid(
    phases,
    load_ctod_func,
    load_hint_func,
    load_model_func,
    thr=0.5,
    cmap=plt.cm.Blues,
    alpha=0.6
):
    """
    Builds a 3x2 grid:
      Rows: Phase I, II, III
      Columns: CTOD (left) and HINT (right)
    """
    nrows, ncols = len(phases), 2
    fig, axes = plt.subplots(
        nrows, ncols,
        figsize=(11, 4.8 * nrows),
        sharex=True, sharey=True,
        gridspec_kw={"wspace": -0.3, "hspace": 0.9}
    )

    def _plot_cm(ax, y_true, proba, title):
        y_pred = (proba >= thr).astype(int)
        cm = confusion_matrix(y_true, y_pred)
        tn, fp, fn, tp = cm.ravel()

        # ---- Metrics ----
        tpr = tp / (tp + fn + 1e-9)
        tnr = tn / (tn + fp + 1e-9)
        prec = tp / (tp + fp + 1e-9)
        acc = accuracy_score(y_true, y_pred)
        bal = 0.5 * (tpr + tnr)

        # ---- Plot matrix ----
        ax.imshow(cm, cmap=cmap, alpha=alpha, interpolation="nearest")
        ax.set_title(title, fontsize=12, fontweight="bold", pad=8)
        ax.set_xlabel("Predicted")
        ax.set_xticks([0, 1]); ax.set_xticklabels(["0 (neg)", "1 (pos)"])
        ax.set_yticks([0, 1]); ax.set_yticklabels(["0 (neg)", "1 (pos)"])

        # Annotate counts
        labels = np.array([["TN", "FP"], ["FN", "TP"]])
        for (r, c), val in np.ndenumerate(cm):
            ax.text(c, r, f"{labels[r, c]}\n{val:,}",
                    ha="center", va="center",
                    fontsize=11, fontweight="bold", color="black")

        # ---- Metrics text ----
        ax.text(
            1.03, 0.5,
            f"Acc={acc:.3f}\n"
            f"Prec={prec:.3f}\n"
            f"Rec={tpr:.3f}\n"
            f"TNR={tnr:.3f}\n"
            f"Bal={bal:.3f}",
            ha="left", va="center", fontsize=10,
            transform=ax.transAxes,
            bbox=dict(boxstyle="round,pad=0.25", facecolor="white", alpha=0.85, linewidth=0)
        )

        # ---- Styling ----
        for spine in ax.spines.values():
            spine.set_visible(False)
        ax.set_xticks(np.arange(-0.5, 2, 1), minor=True)
        ax.set_yticks(np.arange(-0.5, 2, 1), minor=True)
        ax.grid(which="minor", color="white", linestyle="-", linewidth=2)

    # ----- Loop through phases (rows) -----
    for i, phase in enumerate(phases):
        phase_label = phase.replace("Phase", "Phase ")  # ensures consistent spacing

        # Left column → CTOD
        ax_ctod = axes[i, 0]
        X_ctod, y_ctod = load_ctod_func(phase)
        model = load_model_func(phase)
        proba_ctod = model.predict_proba(X_ctod)[:, 1]
        _plot_cm(ax_ctod, y_ctod, proba_ctod, title=f"{phase_label} – CTOD")

        # Right column → HINT
        ax_hint = axes[i, 1]
        X_hint, y_hint = load_hint_func(phase)
        model = load_model_func(phase)
        proba_hint = model.predict_proba(X_hint)[:, 1]
        _plot_cm(ax_hint, y_hint, proba_hint, title=f"{phase_label} – HINT")

    # ---- Global title ----
    fig.suptitle("Confusion Matrices by Phase and Dataset", fontsize=15, fontweight="bold", y=0.98)
    plt.tight_layout()
    plt.show()

# -------- Example call --------
plot_confusion_matrices_ctod_and_hint_grid(
    phases=PHASES,
    load_ctod_func=load_ctod_test,
    load_hint_func=load_hint_test,
    load_model_func=load_hint_model,
    thr=0.5
)

print("\n✅ Confusion matrices grid displayed (rows = phases, columns = datasets).")
```

```python
# ===============================================
# Threshold sweep — HINT→CTOD (top) and HINT→HINT (bottom)
# ===============================================
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Create a 2x3 grid for both datasets
fig, axes = plt.subplots(2, 3, figsize=(15, 10), sharex=True, sharey=True)

# -------- Row 1: HINT → CTOD --------
for i, phase in enumerate(PHASES):
    ax = axes[0, i]
    X_test, y_test = load_ctod_test(phase)
    model = load_hint_model(phase)
    proba = model.predict_proba(X_test)[:, 1]

    ths = np.linspace(0.01, 0.99, 99)
    bal, tnr_list, tpr_list = [], [], []

    for t in ths:
        pred = (proba >= t).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_test, pred).ravel()
        tnr = tn / (tn + fp + 1e-9)
        tpr = tp / (tp + fn + 1e-9)
        tnr_list.append(tnr)
        tpr_list.append(tpr)
        bal.append(0.5 * (tnr + tpr))

    t_bal = float(ths[np.argmax(bal)])

    ax.plot(ths, bal, label="Balanced Accuracy", linewidth=2)
    ax.plot(ths, tnr_list, linestyle="--", label="TNR (Specificity)")
    ax.plot(ths, tpr_list, linestyle="--", label="TPR (Recall)")
    ax.axvline(t_bal, linestyle=":", color="black", label=f"BalAcc-max @ {t_bal:.3f}")
    ax.set_title(f"{phase.upper()}", fontsize=12)
    if i == 0:
        ax.set_ylabel("Score")
        ax.text(-0.25, 0.5, "CTOD", va="center", ha="center",
                rotation=90, fontsize=12, fontweight="bold", transform=ax.transAxes)
    ax.legend(fontsize=8)
    ax.grid(True, linestyle=":", linewidth=0.5)

# -------- Row 2: HINT → HINT --------
for i, phase in enumerate(PHASES):
    ax = axes[1, i]
    X_test, y_test = load_hint_test(phase)
    model = load_hint_model(phase)
    proba = model.predict_proba(X_test)[:, 1]

    ths = np.linspace(0.01, 0.99, 99)
    bal, tnr_list, tpr_list = [], [], []

    for t in ths:
        pred = (proba >= t).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_test, pred).ravel()
        tnr = tn / (tn + fp + 1e-9)
        tpr = tp / (tp + fn + 1e-9)
        tnr_list.append(tnr)
        tpr_list.append(tpr)
        bal.append(0.5 * (tnr + tpr))

    t_bal = float(ths[np.argmax(bal)])

    ax.plot(ths, bal, label="Balanced Accuracy", linewidth=2, color="#1f77b4")
    ax.plot(ths, tnr_list, linestyle="--", label="TNR (Specificity)")
    ax.plot(ths, tpr_list, linestyle="--", label="TPR (Recall)")
    ax.axvline(t_bal, linestyle=":", color="black", label=f"BalAcc-max @ {t_bal:.3f}")
    ax.set_xlabel("Threshold")
    if i == 0:
        ax.set_ylabel("Score")
        ax.text(-0.25, 0.5, "HINT", va="center", ha="center",
                rotation=90, fontsize=12, fontweight="bold", transform=ax.transAxes)
    ax.legend(fontsize=8)
    ax.grid(True, linestyle=":", linewidth=0.5)

# Figure layout
fig.suptitle("Threshold Sweep — HINT→CTOD (top) and HINT→HINT (bottom)",
             fontsize=14, fontweight="bold", y=1.02)
plt.tight_layout()
plt.show()

print("\n✅ Threshold sweep grids displayed for both datasets (CTOD on top, HINT on bottom).")
```

```python
# ===============================================
# Quick metric tables (0.5 vs tuned) — CTOD & HINT side-by-side per phase
# Assumes you already defined:
#   PHASES, load_ctod_test, load_hint_test, load_hint_model,
#   metrics_at_threshold(y_true, proba, thr), pick_threshold_balacc_max(y_true, proba)
# ===============================================
from IPython.display import display, Markdown
import pandas as pd

# --- Helper to build the 0.5 vs tuned table (with Δ row) ---
def summary_table_from(m_default, m_tuned, tuned_thr):
    """Build the comparison DataFrame (0.5 vs tuned) with Δ row."""
    def row_from(m):
        return {
            "Threshold": m["threshold"],
            "Accuracy": m["accuracy"],
            "Precision": m["precision"],
            "Recall (TPR)": m["recall"],
            "Specificity (TNR)": m["tnr"],
            "BalancedAcc": m["bal_acc"],
            "F1": m["f1"],
            "TP": m["cm"]["tp"],
            "FP": m["cm"]["fp"],
            "FN": m["cm"]["fn"],
            "TN": m["cm"]["tn"],
        }

    df = pd.DataFrame([row_from(m_default), row_from(m_tuned)],
                      index=["Default (0.5)", f"Tuned (BalAcc max @ {tuned_thr:.3f})"])

    delta = (df.iloc[1] - df.iloc[0]).to_frame().T
    delta.index = ["Δ (tuned − default)"]
    df = pd.concat([df, delta], axis=0)
    return df

# --- CONFIG for columns to keep/trim ---
SHOW_COUNTS = False  # set True to include TP/FP/FN/TN
BASE_COLS = ["Threshold", "Accuracy", "Precision", "Recall (TPR)", "BalancedAcc", "F1"]
COUNT_COLS = ["TP", "FP", "FN", "TN"]
KEEP_COLS = BASE_COLS + (COUNT_COLS if SHOW_COUNTS else [])

# -------- MAIN --------
combined_phase_tables = {}

for phase in PHASES:
    # ----- CTOD -----
    X_ctod, y_ctod = load_ctod_test(phase)
    model = load_hint_model(phase)
    proba_ctod = model.predict_proba(X_ctod)[:, 1]

    m05_ctod = metrics_at_threshold(y_ctod, proba_ctod, thr=0.5)
    thr_ctod, _ = pick_threshold_balacc_max(y_ctod, proba_ctod)
    mt_ctod = metrics_at_threshold(y_ctod, proba_ctod, thr=thr_ctod)

    df_ctod = summary_table_from(m05_ctod, mt_ctod, thr_ctod)
    # Normalize row labels so CTOD & HINT align when concatenated
    df_ctod.index = ["Default (0.5)", "Tuned", "Δ (tuned − default)"]
    df_ctod = df_ctod[[c for c in KEEP_COLS if c in df_ctod.columns]]

    # ----- HINT -----
    X_hint, y_hint = load_hint_test(phase)
    proba_hint = model.predict_proba(X_hint)[:, 1]  # same phase model

    m05_hint = metrics_at_threshold(y_hint, proba_hint, thr=0.5)
    thr_hint, _ = pick_threshold_balacc_max(y_hint, proba_hint)
    mt_hint = metrics_at_threshold(y_hint, proba_hint, thr=thr_hint)

    df_hint = summary_table_from(m05_hint, mt_hint, thr_hint)
    df_hint.index = ["Default (0.5)", "Tuned", "Δ (tuned − default)"]
    df_hint = df_hint[[c for c in KEEP_COLS if c in df_hint.columns]]

    # ----- Side-by-side table (MultiIndex columns: CTOD | HINT) -----
    df_both = pd.concat({"CTOD": df_ctod, "HINT": df_hint}, axis=1)

    # ----- Display -----
    display(Markdown(f"### Phase: **{phase}** — Quick metric table (CTOD vs HINT)"))
    display(
        df_both.style
            .format("{:.3f}")  # good for all metrics/thresholds; counts (if shown) will render without decimals via int cast below
            .set_caption("Performance summary per threshold (0.5 vs tuned)")
    )

    combined_phase_tables[phase] = df_both

print("\n✅ Built CTOD+HINT side-by-side metric tables for all phases.")
# Access a specific phase table: combined_phase_tables["phase_I"]



```

```python
# ===============================================
# Feature Importance — readable names (all phases)
# Works with SVD-reduced models where booster features are f{idx}
# ===============================================
from pathlib import Path
import re, json, joblib
import pandas as pd
import matplotlib.pyplot as plt
import textwrap
import xgboost as xgb

# ---- Config (edit these to your paths) ----
try:
    MODEL_BASE
except NameError:
    # Fallback if not defined elsewhere
    MODEL_BASE = Path("./xgb_model_package")

try:
    PHASES
except NameError:
    PHASES = ["phase_I", "phase_II", "phase_III"]

TOP_K = 15

# -------- HELPERS --------
def load_hint_model(phase: str):
    """Load model (joblib or JSON) for a phase."""
    mdir = MODEL_BASE / phase
    joblib_path = mdir / "hint_xgb_model.joblib"
    json_path   = mdir / "xgb_model.json"
    if joblib_path.exists():
        return joblib.load(joblib_path)
    if json_path.exists():
        booster = xgb.Booster()
        booster.load_model(str(json_path))
        model = xgb.XGBClassifier()
        model._Booster = booster
        return model
    raise FileNotFoundError(f"No model found for {phase} in {mdir}")

def _strip_leading_index(s: str) -> str:
    """
    Accepts strings like:
      188"pos__criteria__pos__criteria__ may__svd_188"
    and returns:
      pos__criteria__pos__criteria__ may__svd_188
    """
    s = s.strip()
    # If it looks like: 188"...."
    m = re.match(r'^\s*\d+\s*"?(.+?)"?\s*$', s)
    if m:
        return m.group(1)
    # Or maybe it's already just the name
    s = s.strip('"')
    return s

def load_feature_names(phase: str):
    """
    Load readable feature names list for a phase.
    Supports:
      - JSON file that is a list[str]
      - Plain text file with one item per line (optionally prefixed with an index + quote)
    """
    base = MODEL_BASE / phase
    candidates = [
        base / "feature_names_reduced_with_top_original",            # user’s path (no extension)
        base / "feature_names_reduced_with_top_original.json",       # common variant
        base / "feature_names.json",                                 # fallback
        base / "feature_names.txt",                                  # fallback
    ]
    for path in candidates:
        if path.exists():
            try:
                # Try JSON first
                with open(path) as f:
                    data = json.load(f)
                if isinstance(data, list) and all(isinstance(x, str) for x in data):
                    return data
            except Exception:
                # Not JSON; try as plain text
                try:
                    with open(path) as f:
                        lines = [ln.strip() for ln in f if ln.strip()]
                    # Strip "188" prefixes if present
                    cleaned = [_strip_leading_index(ln) for ln in lines]
                    return cleaned
                except Exception:
                    pass
    print(f"[WARN] Feature names file not found for {phase} (looked in {candidates}).")
    return None

def map_f_to_name(fkey: str, feat_names: list[str] | None) -> str:
    """
    Map a booster feature key like 'f280' to readable name via feat_names list.
    If no mapping available, return the original fkey.
    """
    if not (fkey and fkey.startswith("f")) or feat_names is None:
        return fkey
    try:
        idx = int(fkey[1:])
        if 0 <= idx < len(feat_names):
            return feat_names[idx]
    except Exception:
        pass
    return fkey

def _collapse_repeated_subseq(tokens):
    """
    If the full sequence is a repetition of a smaller block (e.g., A,B,A,B),
    collapse it to the smaller block (A,B). Works for any repetition count.
    """
    n = len(tokens)
    if n <= 1:
        return tokens
    for k in range(1, n // 2 + 1):
        if n % k == 0 and tokens == tokens[:k] * (n // k):
            return tokens[:k]
    return tokens

def clean_readable(name: str) -> str:
    """
    Make long technical names compact & human-friendly.
    Removes duplicated blocks like 'POS · ICD POS · ICD' -> 'POS · ICD'.
    """
    orig = str(name)

    # Strip quotes and whitespace
    s = orig.strip().strip('"')

    # Dedup adjacent repeated tokens split by '__' first
    parts = [p for p in s.split("__") if p != ""]
    dedup = []
    for p in parts:
        if not dedup or dedup[-1] != p:
            dedup.append(p)
    s = "__".join(dedup)

    # Pretty block labels
    s = re.sub(r'^pos__', 'POS__', s)
    s = re.sub(r'^sem__', 'SEM__', s)
    s = s.replace("drugs_fusion__", "DRUGS__")
    s = s.replace("diseases_dual__", "DISEASES__")
    s = s.replace("icd__", "ICD__")
    s = s.replace("smiles__", "SMILES__")
    s = s.replace("brief_title__", "TITLE__")
    s = s.replace("description__", "DESC__")

    # Convert __ to " · "
    s = s.replace("__", " · ")

    # Tidy SVD markers: "... _svd_188" -> " · SVD 188"
    s = re.sub(r'\s*_svd[_\- ]?(\d+)', r' · SVD \1', s, flags=re.IGNORECASE)

    # "d29" -> "comp 29"
    s = re.sub(r'(?<![A-Za-z0-9])d(\d+)(?![A-Za-z0-9])', r'comp \1', s)

    # Tokenize on the display separator
    tokens = [t.strip() for t in s.split("·")]
    tokens = [t for t in tokens if t]  # drop empties

    # Remove consecutive duplicates: A, A, B -> A, B
    dedup_tokens = []
    for t in tokens:
        if not dedup_tokens or dedup_tokens[-1] != t:
            dedup_tokens.append(t)

    # Collapse full repeated subsequences: A,B,A,B -> A,B ; A,B,C,A,B,C -> A,B,C
    collapsed = _collapse_repeated_subseq(dedup_tokens)

    # Quote only tiny lowercase tokens (avoid quoting POS/ICD/SEM)
    BLOCK_TAGS = {"POS", "SEM", "ICD", "DRUGS", "DISEASES", "SMILES", "TITLE", "DESC", "SVD"}
    pretty = []
    for t in collapsed:
        tt = t.strip()
        # Normalize spaces around again
        tt = re.sub(r'\s{2,}', ' ', tt)
        if 1 <= len(tt) <= 3 and tt.isalpha() and tt.upper() not in BLOCK_TAGS and tt.islower():
            tt = f'"{tt}"'
        pretty.append(tt)

    out = " · ".join(pretty)
    out = re.sub(r'\s{2,}', ' ', out).strip()
    return out


def wrap_labels(series: pd.Series, width: int = 40) -> list[str]:
    """Wrap long labels for matplotlib barh."""
    wrapped = []
    for x in series:
        wrapped.append("\n".join(textwrap.wrap(str(x), width=width, break_long_words=False)))
    return wrapped

# -------- MAIN LOOP --------
importance_all = {}

for phase in PHASES:
    print(f"\n=== {phase.upper()} ===")
    model = load_hint_model(phase)
    booster = model.get_booster()
    feat_names = load_feature_names(phase)

    # Get importance scores
    gain_scores   = booster.get_score(importance_type="gain")
    cover_scores  = booster.get_score(importance_type="cover")
    weight_scores = booster.get_score(importance_type="weight")

    # Build DataFrame from union of keys (not just gain keys)
    all_keys = set(gain_scores) | set(cover_scores) | set(weight_scores)
    df = pd.DataFrame({
        "Feature": list(all_keys),
    })
    df["Gain"]   = df["Feature"].map(lambda f: gain_scores.get(f, 0.0))
    df["Cover"]  = df["Feature"].map(lambda f: cover_scores.get(f, 0.0))
    df["Weight"] = df["Feature"].map(lambda f: weight_scores.get(f, 0.0))

    # Map to readable names
    df["ReadableFeature_raw"] = df["Feature"].map(lambda f: map_f_to_name(f, feat_names))
    df["ReadableFeature"]     = df["ReadableFeature_raw"].map(clean_readable)

    # Normalize gain (avoid divide-by-zero)
    total_gain = df["Gain"].sum()
    df["Gain_norm"] = df["Gain"] / total_gain if total_gain > 0 else 0.0

    # Sort and keep top-K
    df = df.sort_values("Gain", ascending=False).reset_index(drop=True)
    top = df.head(TOP_K).copy()

    # Plot (horizontal bar, wrapped labels). NOTE: no custom colors per your charting rules.
    plt.figure(figsize=(9, 6))
    labels_wrapped = wrap_labels(top["ReadableFeature"][::-1], width=42)
    plt.barh(range(len(top)), top["Gain_norm"][::-1])
    plt.yticks(range(len(top)), labels_wrapped, fontsize=9)
    plt.xlabel("Normalized Gain")
    plt.title(f"Feature Importance — {phase.upper()}")
    plt.grid(True, axis="x", linestyle=":", linewidth=0.5)
    plt.tight_layout()
    plt.show()

    importance_all[phase] = top[["Feature", "ReadableFeature", "Gain", "Gain_norm", "Cover", "Weight"]]

print("\n✅ Feature-importance plots created with interpretable names for all phases.")

```

```python
# ===============================================
# Aggregate Feature Importance by ORIGINAL COLUMN
# (Renames: sem__criteria -> llm__criteria, sem__smiles -> llm__smiles, num* -> enrollment)
# ===============================================
from pathlib import Path
import re, json, joblib
import pandas as pd
import matplotlib.pyplot as plt
import xgboost as xgb

# ---- Config ----
try:
    MODEL_BASE
except NameError:
    MODEL_BASE = Path("./xgb_model_package")

try:
    PHASES
except NameError:
    PHASES = ["phase_I", "phase_II", "phase_III"]

TOP_K = 12

# ---------- Helpers ----------
def load_hint_model(phase: str):
    mdir = MODEL_BASE / phase
    joblib_path = mdir / "hint_xgb_model.joblib"
    json_path   = mdir / "xgb_model.json"
    if joblib_path.exists():
        return joblib.load(joblib_path)
    booster = xgb.Booster()
    booster.load_model(str(json_path))
    model = xgb.XGBClassifier()
    model._Booster = booster
    return model

def _strip_leading_index(s: str) -> str:
    s = s.strip()
    m = re.match(r'^\s*\d+\s*"?(.+?)"?\s*$', s)
    if m:
        return m.group(1)
    return s.strip('"')

def load_feature_names(phase: str):
    base = MODEL_BASE / phase
    candidates = [
        base / "feature_names_reduced_with_top_original",
        base / "feature_names_reduced_with_top_original.json",
        base / "feature_names.json",
        base / "feature_names.txt",
    ]
    for path in candidates:
        if path.exists():
            try:
                with open(path) as f:
                    data = json.load(f)
                if isinstance(data, list) and all(isinstance(x, str) for x in data):
                    return data
            except Exception:
                pass
            try:
                with open(path) as f:
                    lines = [ln.strip() for ln in f if ln.strip()]
                return [_strip_leading_index(ln) for ln in lines]
            except Exception:
                pass
    print(f"[WARN] Feature names file not found for {phase}.")
    return None

def map_f_to_name(fkey: str, feat_names: list[str] | None) -> str:
    if not (fkey and fkey.startswith("f")) or feat_names is None:
        return fkey
    try:
        idx = int(fkey[1:])
        if 0 <= idx < len(feat_names):
            return feat_names[idx]
    except Exception:
        pass
    return fkey

# ---------- Origin column extraction ----------
HEAD_WIDTH = {
    "pos": 2,
    "sem": 2,
    "diseases_dual": 1,
    "drugs_fusion": 1,
    "icd": 1,
    "smiles": 1,
    "brief_title": 1,
    "description": 1,
    "title": 1,
    "desc": 1,
    "num": 1,
}

SVD_TOKENS_RE = re.compile(r'^(svd(_component)?(_?\d+)?)$', re.IGNORECASE)

def _collapse_adjacent_dupes(parts):
    out = []
    for p in parts:
        if not out or out[-1] != p:
            out.append(p)
    return out

def _collapse_full_repeat(tokens):
    n = len(tokens)
    if n <= 1:
        return tokens
    for k in range(1, n // 2 + 1):
        if n % k == 0 and tokens == tokens[:k] * (n // k):
            return tokens[:k]
    return tokens

def extract_origin_column(raw_name: str) -> str:
    """
    Extract origin column (e.g., pos__criteria, sem__criteria, drugs_fusion, etc.)
    and rename:
      sem__criteria -> llm__criteria
      sem__smiles   -> llm__smiles
      num*          -> enrollment
    """
    s = _strip_leading_index(str(raw_name)).strip().strip('"')
    parts = [p for p in s.split("__") if p]
    parts = _collapse_adjacent_dupes(parts)
    parts = _collapse_full_repeat(parts)

    cleaned = []
    for p in parts:
        if SVD_TOKENS_RE.match(p):
            continue
        if re.fullmatch(r'[dD]\d+', p):
            continue
        cleaned.append(p)
    parts = cleaned

    if not parts:
        return str(raw_name)

    head = parts[0].lower()
    if head in HEAD_WIDTH:
        take = HEAD_WIDTH[head]
        take = min(take, len(parts))
        origin = "__".join(parts[:take])
    else:
        origin = parts[0]

    # --- Rename for visualization ---
    if origin.startswith("sem__criteria"):
        origin = "llm__criteria"
    elif origin.startswith("sem__smiles"):
        origin = "llm__smiles"
    elif origin.startswith("num"):
        origin = "enrollment"

    return origin

# ---------- Main aggregation ----------
agg_by_phase = {}

for phase in PHASES:
    print(f"\n=== {phase.upper()} — aggregate by origin column ===")
    model = load_hint_model(phase)
    booster = model.get_booster()
    feat_names = load_feature_names(phase)

    gain_scores   = booster.get_score(importance_type="gain")
    cover_scores  = booster.get_score(importance_type="cover")
    weight_scores = booster.get_score(importance_type="weight")

    all_keys = set(gain_scores) | set(cover_scores) | set(weight_scores)
    df = pd.DataFrame({"BoosterKey": list(all_keys)})
    df["Gain"]   = df["BoosterKey"].map(lambda f: gain_scores.get(f, 0.0))
    df["Cover"]  = df["BoosterKey"].map(lambda f: cover_scores.get(f, 0.0))
    df["Weight"] = df["BoosterKey"].map(lambda f: weight_scores.get(f, 0.0))

    df["RawName"] = df["BoosterKey"].map(lambda f: map_f_to_name(f, feat_names))
    df["OriginColumn"] = df["RawName"].map(extract_origin_column)

    agg = df.groupby("OriginColumn", as_index=False).agg(
        Gain=("Gain", "sum"),
        Cover=("Cover", "sum"),
        Weight=("Weight", "sum"),
        Features=("BoosterKey", "count"),
    ).sort_values("Gain", ascending=False)

    total_gain = agg["Gain"].sum()
    agg["Gain_pct"] = (agg["Gain"] / total_gain * 100.0) if total_gain > 0 else 0.0

    out_dir = MODEL_BASE / phase
    out_dir.mkdir(parents=True, exist_ok=True)
    out_csv = out_dir / "feature_importance_by_origin.csv"
    agg.to_csv(out_csv, index=False)
    print(f"Saved: {out_csv}")

    # --- Plot ---
    top = agg.head(TOP_K).copy()
    plt.figure(figsize=(8, 8))
    plt.barh(range(len(top)), top["Gain_pct"][::-1])
    plt.yticks(range(len(top)), top["OriginColumn"][::-1], fontsize=19)
    plt.xlabel("Total Gain (%)",fontsize = 19)
    plt.title(f"Feature Importance by Origin — {phase.upper()}", fontsize = 19)
    plt.grid(True, axis="x", linestyle=":", linewidth=0.5)
    plt.tight_layout()
    plt.show()

    agg_by_phase[phase] = agg

print("\n✅ Aggregated importance by original column for all phases (with llm__criteria, llm__smiles, and enrollment labels).")

```
